```python
# Problem Statement
'''
Given a list of numbers i.e., [1,2,4,-5,7,9,3,2], make another list contains
all the items in sorted order from min to max. Your result will be another
list like [-5,1,2,2,3,4,7,9]
'''
L = [1,2,4,-5,7,9,3,2]
for j in range(len(L)):
    m = L[j]
    idx = j
    c = j
    for i in range(j, len(L)):
        if L[i] < m:
            m = L[i]
            idx = c
        c += 1
    temp = L[j]
    L[j] = m
    L[idx] = temp
print(L)
        
```

**Deep copy** and **Shallow copy** concept

- **Deep copy** stores copies of the object's value.
- **Shallow copy** reflects changes made to the new/copied object in the original object.
- **Deep copy** doesn't reflect changes made to the new/copied object in teh original object.
- **Shallow Copy** stores the copy of the original object and pointsteh references to tehh objects.

When we call a function with a value passed as argument; it will produce a shallow copy of object passed to variable.i.e., value is passed as reference as in above case.

```python
# Shallow copy Example
L1 = [1,2,3]
L2 = L1
L2[0] = 12
print(L1)
```

## Practice of Functions

```python
'''
Given a list of numbers i.e., [1,2,4,-5,7,9,3,2], make another list contains
all the items in sorted order from min to max. Your result will be another
list like [-5,1,2,2,3,4,7,9]
'''
def findMin(L, startIndex):
    '''
    will return a minimum function and its index in the list
    '''
    m = L[startIndex]
    idx = startIndex
    for i in range(startIndex, len(L)):
        x = L[i]
        if x < m:
            m = x
            idx = i
        i += 1
    return m, idx
```

```python
a, b = findMin([2,6,9,1,4,5,2],4)
print(a,b)
```

```python
def swapValues(L, idx1, idx2):
    temp = L[idx1]
    L[idx1] = L[idx2]
    L[idx2] = temp
    return L
```

```python
L = [8,4,3,2,9]
print(swapValues(L, 1, 4))
```

```python
def checkIfNotnumeric(L):
    for x in L:
        if not(isinstance(x, (int, float))):
            return False
    return True
```

```python
def sortList(L):
    '''
    Sort a Numeric valued list and return it.
    '''
    if not(checkIfNotnumeric(L)):
        print('Error: List does not contain numeric values.')
        return
    else:
        c = 0 
        for x in L:
            m, idx = findMin(L, c)
            L = swapValues(L, c, idx)
            c += 1
    return L
```

```python
print([3,2,8,-1,4,-9,5])
print(sortList([3,2,8,-1,4,-9,5]))
```

# Data Structures

- List:  `(      Ordered            Changable                      Duplicates)`
- Tuple: `(    Ordered            Unchangable                    Duplicates)`
- Set:   `(      Unordered          addable\removeable             no duplicates)`

|           List        |  Tuple              | Set                       |  Dictionary                       |
|:----------------------|:--------------------|:--------------------------|:----------------------------------|

|L2=L.copy              |T2=T                 |S2=S.copy()                |D2=D.copy()                        |

```python
# Problem statement
'''
Lets say you are a teacher and you have different students record containing ID of student, 
marks list in each suject; where different students have taken different number of subjects. 
All these records are in hard copy. You want to enter all the data in computer and want to 
compute the average marks of each student and display.
'''

def getDataFromUser():
    D = {}
    while True:
        studentID = input("Enter student ID: ")
        marksList = input("Enter marks by comma seprated values: ")
        if studentID in D:
            print(studentID, "is already inserted")
        else:
            D[studentID] = marksList.split(',')
        moreStudents = input('Enter "No" to quit insertion: ')
        if moreStudents.lower() == 'no':
            return D
```

```python
studentData = getDataFromUser()
```

```python
print(studentData)
```

```python
print(getDataFromUser())
```

```python
def getAvgMarks(D):
    avgMarks = {}
    for x in D:
        L = D[x]
        s = 0
        for marks in L:
            s += int(marks)
        avgMarks[x] = s/len(L)
    return avgMarks
```

```python
avgM = getAvgMarks(studentData)
```

```python
print(avgM)
```

```python
for x in avgM:
    print(f"Student {x} got {avgM[x]} as average marks.")
```
